**Expense Tracker**

```python
class Expense:
    def __init__(self, name, category, amount):
        self.name = name
        self.category = category
        self.amount = amount

    def __repr__(self):
        return f"Expense: Name: {self.name} Category: {self.category} Amount: â‚¹{self.amount:.2f}"
```

```python
def main():
    print(f"ğŸ¯ Running...")
    expense_file_path = "expense.csv"
    budget_file_path = 'budget.csv'
    # Get user input of their expense
    expense = get_user_expense()
    # Write it to a file
    with open('budget.csv', 'r') as file:
        remaining_budget = file.read()
        if expense.amount > float(remaining_budget):
            print(f'You are moving out from your this month budget.\nğŸ’°Remaining Budget: {remaining_budget}')
            return
    save_expense(expense, expense_file_path)
    # Read the file and summarize all the expenses
    summarize_expense(expense_file_path)
    with open('budget.csv', 'w') as file:
        budget = float(remaining_budget) - expense.amount
        file.write(str(budget))
    print(f"Remaining Budget for this month {budget}")



def get_user_expense():
    print(f"ğŸ¯ Get user expense...")
    expense_name = input("Enter expense name: ").strip()
    expense_amount = float(input("Enter expense amount: "))
    expense_categories = [
        "ğŸ• Food", "ğŸ¡ Home", "ğŸ’¼ Work", "ğŸ‰ Fun", "ğŸ—‚ï¸ Miscellaneous"
    ]
    while True:
        print("Select a category: ")
        for i, name in enumerate(expense_categories):
            print(i + 1, ".", name)
        value_range = f"[1 - {len(expense_categories)}]: "
        try:
            select_index = int(input(f"Enter a category number {value_range}")) - 1
            if select_index in range(len(expense_categories)):
                new_expense = Expense(expense_name, expense_categories[select_index], expense_amount)
                return new_expense
            else:
                print("Invalid category. Please enter a valid one from list")
        except ValueError:
            print("Enter a number from available categories. Integer type cannot have string input")


def save_expense(expense, expense_file_path):

    print(f"ğŸ¯ Saving your expense...{expense}")

    with open(expense_file_path, 'a', encoding='utf-8') as file:
        file.write(f"{expense.name}, {expense.category}, {expense.amount}\n")


def summarize_expense(expense_file_path):
    print(f"ğŸ¯ Getting summary of your expenses...")
    expenses = []
    with open(expense_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        for line in lines:
            expense_name, expense_category, expense_amount = line.strip().split(", ")
            line_expense = Expense(expense_name, expense_category, float(expense_amount))
            expenses.append(line_expense)
    amount_by_category = {}
    for expense in expenses:
        key = expense.category
        if key in amount_by_category:
            amount_by_category[key] += expense.amount
        else:
            amount_by_category[key] = expense.amount
    for key, amount in amount_by_category.items():
        print(f"Category: {key}: â‚¹{amount:.2f}")

    total_spent = sum([expense.amount for expense in expenses])
    print(f"ğŸ’¹ You have spent â‚¹{total_spent:.2f} this month!")


if __name__ == "__main__":
    main()


```
